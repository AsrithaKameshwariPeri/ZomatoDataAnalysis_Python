```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the restaurant data
# Replace 'Cuisines' with the actual column name containing cuisine information

# Split cuisines and count occurrences
cuisine_counts = df['Cuisines'].str.split(', ').explode().value_counts().head(10)

# Plotting with seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values, palette='viridis')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top 10 Cuisines')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation:From the above plott the top 3 cuisines are North Indian ,Chinese,Fast Food through Zomato app

```python
#count of restaurants with online delivery
plt.figure(figsize=(8, 6))
sns.countplot(x='Has Online delivery', data=df, palette='Set3')
plt.xlabel('Has Online Delivery')
plt.ylabel('Count')
plt.title('Count of Restaurants with Online Delivery')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Has Online delivery', data=df, palette='Set3')

# Display count numbers on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.xlabel('Has Online Delivery')
plt.ylabel('Count')
plt.title('Count of Restaurants with Online Delivery')
plt.show()
```

Observation: 
1)The count of restaurants with online delivery are 2451.
2)The count of restaurants without online delivery are 7100.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check the exact column names in your dataframe
print(df.columns)

# Use the correct column name in the plot
plt.figure(figsize=(12, 8))
ax = sns.countplot(x='Country', data=final_df, palette='plasma')
plt.title('Count of Restaurants in Each Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Annotate each bar with its count value
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

Observation:
1.From the above plott the count of restaurants is highest in India.

```python
#Cuisines vs number of restaurants
plt.figure(figsize=(14, 8))
top_cuisines = df['Cuisines'].value_counts().head(10)  # Select top 10 cuisines
bar_plot = sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette='viridis')

for index, value in enumerate(top_cuisines.values):
    bar_plot.text(index, value, str(value), ha='center', va='bottom', fontsize=10)

plt.title('Top 10 Cuisines vs. no.of.restaurants')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

Observation:
1)From the above bar_plot it is clear that maximum number of restaurants have North Indian cuisine with them.
